```python
import pandas as pd

train_df = pd.read_csv("titanic_data/train.csv")
test_df = pd.read_csv("titanic_data/test.csv")

print(train_df.head())
print(train_df.shape)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler

# Fill missing values
train_df['Age'] = train_df['Age'].fillna(train_df['Age'].median())
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].median())

train_df['Embarked'] = train_df['Embarked'].fillna(
    train_df['Embarked'].mode()[0]
)

# Check if Fare exists in test data and fill missing values
if 'Fare' in test_df.columns:
    test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].median())

# Encode categorical columns
le_sex = LabelEncoder()
train_df['Sex'] = le_sex.fit_transform(train_df['Sex'])
test_df['Sex'] = le_sex.transform(test_df['Sex'])

le_embarked = LabelEncoder()
train_df['Embarked'] = le_embarked.fit_transform(train_df['Embarked'])
test_df['Embarked'] = le_embarked.transform(test_df['Embarked'])

# Drop unnecessary columns
train_df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)
test_df.drop(['PassengerId', 'Name', 'Ticket'], axis=1, inplace=True)

# Features and target
X = train_df.drop('Survived', axis=1)
y = train_df['Survived']

# Feature scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_test_scaled = scaler.transform(test_df)

print("Preprocessing completed successfully!")
print("X shape:", X.shape)
print("y shape:", y.shape)
```
