```python
scores = list(map(int, input("Enter scores separated by spaces: ").split()))

largest = float('-inf')
second_largest = float('-inf')

for score in scores:
    if score > largest:
        second_largest = largest
        largest = score
    elif score > second_largest and score != largest:
        second_largest = score

if second_largest == float('-inf'):
    print("There is no second largest score.")
else:
    print("Second Largest Score:", second_largest)
```

```python
text = input("Enter a string: ")

for char in text:
    if text.count(char) == 1:
        print("First unique character:", char)
        break
else:
    print("No unique character found.")
```

```python
n = int(input("Enter the number of elements: "))

lst = []
for i in range(n):
    lst.append(int(input(f"Enter element {i+1}: ")))

k = int(input("Enter K value: "))

k = k % n  # Handle K greater than list size

rotated = lst[-k:] + lst[:-k]

print("Rotated List:", rotated)
```

```python
nums = list(map(int, input("Enter numbers separated by spaces: ").split()))

even = []
odd = []

for num in nums:
    if num % 2 == 0:
        even.append(num)
    else:
        odd.append(num)

result = even + odd

print("Rearranged List:", result)
```

```python
def nested_sum(lst):
    total = 0

    for item in lst:
        if isinstance(item, list):
            total += nested_sum(item)
        else:
            total += item

    return total

data = [1, [2, 3], [4, [5, 6]]]

print("Sum =", nested_sum(data))
```

```python
for num in range(1, 101):
    digits = [int(d) for d in str(num)]

    digit_sum = sum(digits)

    digit_product = 1
    for d in digits:
        digit_product *= d

    if digit_sum == digit_product:
        print(num, end=" ")
```

```python
arr = list(map(int, input("Enter array elements: ").split()))
target = int(input("Enter target sum: "))

for i in range(len(arr)):
    for j in range(i + 1, len(arr)):
        if arr[i] + arr[j] == target:
            print((arr[i], arr[j]))
```
